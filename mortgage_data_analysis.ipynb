```python
# Getting some baseline stats for the columns with missing values
cols_with_missing = df[['borrower_income', 'credit_score', 'dti_ratio']]

cols_with_missing.describe()
```

```python
# Plotting the distributions of each of the missing columns to make sure I fill missing values correctly
for col in cols_with_missing:
    plt.figure(figsize = (8, 4))
    cols_with_missing[col].hist(bins=30, edgecolor='black')
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.tight_layout()
    plt.show()
    
```

Given that each of the variables follow a normal distribution, it is fair to use the mean value of the column for imputation on the missing values.

```python
# Filling each column with the mean of the column
for col in cols_with_missing.columns:
    df[col] = df[col].fillna(df[col].mean())

df.isna().sum()
```

```python
# Pulling some columns I want to get aggregate values for.
cols_to_avg = ['loan_amount', 'dti_ratio', 'ltv_ratio', 'credit_score', 'property_value', 'interest_rate']

# Pulling some columns to aggregate based on.
cols_to_group_by = ['state', 'loan_term', 'loan_status']

# Function to create csv files with aggregate stats for future plotting
for grouped_col in cols_to_group_by:
    df_for_agg_cols = pd.DataFrame() # New dataframe to hold the aggregate values for the specific groupby column.
    
    # Loop through each column to grab the average for a given column creating a new df with the agg stats for that column, appending to the main df for that group.
    for c in cols_to_avg:
        agg_stat_df = df.groupby(grouped_col)[c].mean().sort_index().round(2)
        df_for_agg_cols[f"{c}_avg_by_{grouped_col}"] = agg_stat_df.values
    
    # Ensuring the index is set
    df_for_agg_cols.index = agg_stat_df.index

    # Creating a csv for the aggregate stats
    df_for_agg_cols.to_csv(f"metrics_grouped_by_{grouped_col}.csv", index=True)



```

```python
# New dfs with the aggregate stats
state_stats_df = pd.read_csv("metrics_grouped_by_state.csv", index_col="state")

loan_term_stats_df = pd.read_csv("metrics_grouped_by_loan_term.csv", index_col = "loan_term")

loan_status_stats_df = pd.read_csv("metrics_grouped_by_loan_status.csv", index_col="loan_status")
```
